```python
import pandas as pd
import json

with open('../data/tracking_age.json') as f:
    data = json.load(f)

tracking_df = pd.DataFrame(data)[['Country', 'EarliestStreamingAge']]
tracking_df = tracking_df.rename(columns={'EarliestStreamingAge': 'tracking_age'})
```

```python
tracking_df.to_csv('../output/tracking_age.csv', index=False)
```
